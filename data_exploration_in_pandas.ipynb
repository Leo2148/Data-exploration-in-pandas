# Basic visualizations

```python
df = pop_df3.copy()
```

Analize distributions of each attribute

```python
for col in df.columns:
    if col != 'Balance':
        df.plot(x=col, y='Balance', kind="scatter")
        plt.title(col+" and Balance")
        plt.show()
```

```python
# plot histogram and boxplot for each attribute
for col in df.columns:
    df.hist(col)
    plt.show()
    df.boxplot(col)
    plt.show()
```
